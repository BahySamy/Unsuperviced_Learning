```python
def comparing_clusters(dataset, nclusters, epsoln, min_points):
    
    idx_kmean = KMeans(n_clusters = nclusters).fit_predict(dataset.to_numpy())
    idx_agg = AgglomerativeClustering(n_clusters = nclusters).fit_predict(dataset.to_numpy())
    idx_dbscan = DBSCAN(eps = epsoln, min_samples = min_points).fit_predict(dataset.to_numpy())
    plt.figure(figsize = (15,4)) 
    ax = plt.subplot(1,4,1)
    ax.title.set_text('Original Data')
    sns.scatterplot(data = dataset, x='x', y='y');
    ax = plt.subplot(1,4,2)
    ax.title.set_text('Kmeans')
    sns.scatterplot(data = dataset, x='x', y='y', hue=idx_kmean, ax = ax);
    ax = plt.subplot(1,4,3)
    ax.title.set_text('Agglomerative')
    sns.scatterplot(data = dataset, x='x', y='y', hue=idx_agg, ax = ax);
    ax = plt.subplot(1,4,4)
    ax.title.set_text('DBSCAN')
    sns.scatterplot(data = dataset, x='x', y='y', hue=idx_dbscan, ax = ax)
    
    
```
